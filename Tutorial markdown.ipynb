## 1. Como fazer 'escape' de comando?   
Use três crases seguidas  
Finalize com três crases seguidas    
Ex:    
Código: 

```
# Jogo da velha e uma frase
```
Resultado:  
### Três jogos da velha e uma frase

Eu usarei o método das três crases a partir de agora para mostrar os comandos

```
Soma:  
$x + y$  
Subitração:   
$x - y$   
Multiplicação:  
$x * y$  
$x \times y$  
Divisão:  
$x / y$  
$x \div y$  
$\frac  {x}{y}$  
```

#### Resultado:  
Soma:  
$x + y$  
Subitração:   
$x - y$   
Multiplicação:  
$x * y$  
$x \times y$  
Divisão:  
$x / y$  
$x \div y$  
$\frac  {x}{y}$  

#### Símbolos:
Código:
```
$\pi$    
$\approx$  
$\leq$  
$\geq $  
```
#### Resultado: 
$\pi$    
$\approx$  
$\leq$  
$\geq $  

Mais informações:
[LaTeX Wiki](https://en.wikibooks.org/wiki/LaTeX/Mathematics)
